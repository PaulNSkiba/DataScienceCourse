Напишіть функції, що:

* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random, math

#генеруэмо тисячу чисел в діапазоні від 0 до 10 (!!!!!!!!!!!!!!!!!!!!!!??????????????????????????)
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

def getMinMax(arr, isMin):
    getMinMax = arr[0]
    for i in arr:
        if getMinMax > i and not isMin:
            getMinMax = i
        if getMinMax < i and isMin:
            getMinMax = i    
    return getMinMax

print('Min value: ', getMinMax(list_with_random_numbers, False))        
print('Max value: ', getMinMax(list_with_random_numbers, True))  

def getAvgValue(arr):
    sm = 0
    for i in arr:
        sm+=i
    return sm/len(arr)

print('Avg value: ', getAvgValue(list_with_random_numbers))

def getMediana(arr):
    arrCopy = arr[:]
    arrCopy.sort()
    return arr[500]
                
print('Mediana value: ', getMediana(list_with_random_numbers))    

number = 63
def getNumberFrequency(arr, number):
    numbers = dict()
    for i in arr:
        if i in numbers:
            numbers[i]+=1
        else:
            numbers[i]=1
    return numbers[number]        

print(f'Number frequency of {number}: ', getNumberFrequency(list_with_random_numbers, number))    

def getDispersion(arr):
    avg = getAvgValue(arr)
    avgDispersion = 0
    for i in arr:
        avgDispersion += (i - avg)**2
    return  math.sqrt(avgDispersion/(len(arr) - 1))

print('Dispersion: ', getDispersion(list_with_random_numbers))
```
