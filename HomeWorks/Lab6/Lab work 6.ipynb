```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns

np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})

fig, axs = plt.subplots(3, 1, figsize=(20, 10))
#print(df['a'])
axs[0].plot(df['a'], '-', label='Line A')
axs[0].legend(loc='lower right', labelspacing=1.2)
axs[1].plot(df['b'], '-', label='Line B')
axs[1].legend(loc='center left', labelspacing=1.2)
axs[2].plot(df['c'], '-', label='Line C')
axs[2].legend(loc='upper right', labelspacing=1.2)
```

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns

np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})

fig, axes = plt.subplots()

axes.bar(range(len(df.a)), df.a, alpha=0.5, color='#ff6347', width=.75)
plt.legend('a')
```

Створити візуалізацію, аналогічно рисунку 
- добавте всі підписи та правильний маркер

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns

np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})

fig, axes = plt.subplots()

plt.xticks(range(len(df.a)))
axes.plot(df.a, '-^', label='Line A', linewidth=2)
axes.legend(loc='upper right', labelspacing=1.2)
axes.set_title('This is the Title')
axes.set_xlabel('This is the X Axis', size=20)
axes.set_ylabel('This is the Y Axis')
```
